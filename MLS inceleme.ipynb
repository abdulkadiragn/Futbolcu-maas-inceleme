```python
import pandas as pd
```

```python
#veriyi okuma
data = pd.read_csv("mls-salaries.csv")
```

```python
#ilk 20 veri
data.head(n=20)
```

```python
#kaç tane veri olduğunu yazar
len(data.index)
```

```python
#tüm maaşların ortalaması
data["base_salary"].mean()
```

```python
#en yüksek maaş
data["base_salary"].max()
```

```python
#en yüksek tazminata sahip futbolcunun klübü
data[data["guaranteed_compensation"] == data["guaranteed_compensation"].max()]["club"].iloc[0]
```

```python
#loyd soyadlı oyuncunun maaşını bul
data[data["last_name"] == "Loyd"]["base_salary"].iloc[0]
```

```python
#ATL takımında kaç tane D var bul
data[(data["position"] == "D") & (data["club"] == "ATL")]
```

```python
#ATL Defans oyuncularının ortalama maaşları
data[(data["position"] == "D") & (data["club"] == "ATL")]["base_salary"].mean()
```

```python
#mevkilere göre futbolcuları grupla
#mevkilerde ort maaşı bul
data.groupby("position").mean()
```

```python
#kaç farklı mevki vardır ? 
len(data.groupby("position"))
```

```python
#kaç farklı mevki var 2.çözüm
data["position"].nunique()
```

```python
#her mevkide kaç oyuncu var
data["position"].value_counts()
```

```python
#her takımda kaç oyuncu bulunmakta
data["club"].value_counts()
```

```python
#isimlerinin içinde "an" geçen futbolcular
data[data["last_name"].str.contains('an')]
```

```python
#maaşı 77004.0'den fazla olan oyuncular
data[data["base_salary"] > 77004.0]
```

```python
#bonservisi 450000.00'dan fazla olanlar
data[data["guaranteed_compensation"] > 450000.00]
```
